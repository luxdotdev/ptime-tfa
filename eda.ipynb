```python
# Load cleaned data from collection script
import collect_data

df = collect_data.df

df.head()
```

```python
from statsmodels.graphics.mosaicplot import mosaic
mosaic(df, ['victim_role', 'attacker_team_won'], title = 'Fight Wins by First Death Role')
```

```python
# Subset for FIU first bloods
fiu_kill = df[df['attacker_team'] == 'Team 1']

# Filter out subs
fiu_kill = fiu_kill[fiu_kill['attacker_name'].isin(['Spingar','TWEENER','Asta','Georan','H1dd3n', 'YAI'])]

mosaic(fiu_kill, ['victim_role', 'attacker_team_won'], title = 'Fight Wins by FIU First Kill Role')
mosaic(fiu_kill, ['attacker_name', 'attacker_team_won'], title = 'Fight Wins by First Blood Getter')
```
