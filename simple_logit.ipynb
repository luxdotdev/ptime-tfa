```python
import collect_data

dat = collect_data.df

dat.head(10)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Prepare features and target variable
X = dat[['victim_role']]
y = dat['attacker_team_won']

# Encode categorical feature
encoder = OneHotEncoder()
X_encoded = encoder.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size = 0.2, random_state = 69)

# Initialize and train a LogisticRegression model
model = LogisticRegression()
model.fit(X_train, y_train)

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

# Predict probabilities on the test set
y_pred_prob = model.predict_proba(X_test)[:, 1]

# Evaluate model performance
accuracy = accuracy_score(y_test, y_pred_prob > 0.5)
print("Accuracy:", accuracy)
```

1. **"Damage" Role**:
   - For each instance where the victim's role is categorized as "Damage", the log-odds of the attacker team winning decrease by approximately 18.7% compared to the reference category.
   - In terms of probabilities, this translates to a decrease in the likelihood of the attacker team winning by approximately 15.8%.

2. **"Support" Role**:
   - For each instance where the victim's role is categorized as "Support", the log-odds of the attacker team winning decrease by a negligible amount (0.02%) compared to the reference category.
   - In terms of probabilities, this translates to a negligible change in the likelihood of the attacker team winning.

3. **"Tank" Role**:
   - For each instance where the victim's role is categorized as "Tank", the log-odds of the attacker team winning increase by approximately 17.4% compared to the reference category.
   - In terms of probabilities, this translates to an increase in the likelihood of the attacker team winning by approximately 18.0%.

According to this model output, focusing on eliminating the opposing Tank first yields the greatest increase in the likelihood of the attacker team winning team fights on average.
